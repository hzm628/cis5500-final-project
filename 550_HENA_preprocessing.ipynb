```python
# Keeping only necessary columns and renaming
cities_data = cities_data[["Country name EN", "ASCII Name", "Population", "Coordinates" ]]
cities_data = cities_data.rename(columns={"ASCII Name" : "City"})
cities_data = cities_data.rename(columns={"Population" : "City_Population"})
cities_data = cities_data.rename(columns={"Country name EN" : "Country"})
```

```python
cities_data = cities_data.dropna()
```

```python
cities_data.drop(cities_data[cities_data['City_Population'] == 0].index, inplace = True)
```

```python
# Separating coordinates into Longitude and Latitude

# Regex pattern to match latitude
pattern = r'(.*),'

cities_data["City_Latitude"] = cities_data["Coordinates"].str.extract(pattern, flags=re.IGNORECASE)

# Regex pattern to match longitude
pattern = r', (.*)'

cities_data["City_Longitude"] = cities_data["Coordinates"].str.extract(pattern, flags=re.IGNORECASE)

# Drop coordinates
cities_data = cities_data.drop(columns = "Coordinates")
```

```python
# Remove apostrophe
cities_data['City'] = cities_data['City'].str.replace(r"'", "", regex=True)
```

```python
cities_data = cities_data.astype({'City_Latitude' : float})
cities_data = cities_data.astype({'City_Longitude' : float})
```

```python
cities_data.sort_values(["City_Population"], ascending=[False], inplace = True)
```

```python
cities_data.drop_duplicates(subset = ['City', 'Country'], inplace = True, keep = 'first')
```

```python
ids = cities_data["City"]
cities_data[ids.isin(ids[ids.duplicated()])].sort_values("City")
```

```python
cities_data.sort_values(["City", "City_Population"], ascending=[True, False], inplace = True)
```

```python
cities_data = cities_data.reset_index(drop = True)
```

```python
countries_data = countries_data.drop(columns= {'Density\n(P/Km2)', 'Abbreviation','Armed Forces size', 'Calling Code', 'Capital/Major City', 'Currency-Code', 'Maternal mortality ratio'})
```

```python
# Remove commas
countries_data['Agricultural Land( %)'] = countries_data['Agricultural Land( %)'].str.replace(r",", "", regex=True)
countries_data["Land Area(Km2)"] = countries_data["Land Area(Km2)"].str.replace(r",", "", regex=True)
countries_data["Co2-Emissions"] = countries_data["Co2-Emissions"].str.replace(r",", "", regex=True)
countries_data["CPI"] = countries_data["CPI"].str.replace(r",", "", regex=True)
countries_data["CPI Change (%)"] = countries_data["CPI Change (%)"].str.replace(r",", "", regex=True)
countries_data['Forested Area (%)'] = countries_data['Forested Area (%)'].str.replace(r",", "", regex=True)
countries_data['Gasoline Price'] = countries_data['Gasoline Price'].str.replace(r",", "", regex=True)
countries_data['GDP'] = countries_data['GDP'].str.replace(r",", "", regex=True)
countries_data['Gross primary education enrollment (%)'] = countries_data['Gross primary education enrollment (%)'].str.replace(r",", "", regex=True)
countries_data['Gross tertiary education enrollment (%)'] = countries_data['Gross tertiary education enrollment (%)'].str.replace(r",", "", regex=True)
countries_data['Minimum wage'] = countries_data['Minimum wage'].str.replace(r",", "", regex=True)
countries_data['Out of pocket health expenditure'] = countries_data['Out of pocket health expenditure'].str.replace(r",", "", regex=True)
countries_data['Population: Labor force participation (%)'] = countries_data['Population: Labor force participation (%)'].str.replace(r",", "", regex=True)
countries_data['Tax revenue (%)'] = countries_data['Tax revenue (%)'].str.replace(r",", "", regex=True)
countries_data['Total tax rate'] = countries_data['Total tax rate'].str.replace(r",", "", regex=True)
countries_data['Unemployment rate'] = countries_data['Unemployment rate'].str.replace(r",", "", regex=True)
countries_data["Population"] = countries_data["Population"].str.replace(r",", "", regex=True)
countries_data["Urban_population"] = countries_data["Urban_population"].str.replace(r",", "", regex=True)
```

```python
# Remove % and $
countries_data['Agricultural Land( %)'] = countries_data['Agricultural Land( %)'].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'Agricultural Land( %)' : float})
countries_data['Agricultural Land( %)'] = countries_data['Agricultural Land( %)']/100

countries_data["CPI Change (%)"] = countries_data["CPI Change (%)"].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'CPI Change (%)' : float})
countries_data['CPI Change (%)'] = countries_data['CPI Change (%)']/100

countries_data['Forested Area (%)'] = countries_data['Forested Area (%)'].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'Forested Area (%)' : float})
countries_data['Forested Area (%)'] = countries_data['Forested Area (%)']/100

countries_data['Gross primary education enrollment (%)'] = countries_data['Gross primary education enrollment (%)'].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'Gross primary education enrollment (%)' : float})
countries_data['Gross primary education enrollment (%)'] = countries_data['Gross primary education enrollment (%)']/100

countries_data['Gross tertiary education enrollment (%)'] = countries_data['Gross tertiary education enrollment (%)'].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'Gross tertiary education enrollment (%)' : float})
countries_data['Gross tertiary education enrollment (%)'] = countries_data['Gross tertiary education enrollment (%)']/100

countries_data['Population: Labor force participation (%)'] = countries_data['Population: Labor force participation (%)'].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'Population: Labor force participation (%)' : float})
countries_data['Population: Labor force participation (%)'] = countries_data['Population: Labor force participation (%)']/100

countries_data['Tax revenue (%)'] = countries_data['Tax revenue (%)'].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'Tax revenue (%)' : float})
countries_data['Tax revenue (%)'] = countries_data['Tax revenue (%)']/100

countries_data['Total tax rate'] = countries_data['Total tax rate'].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'Total tax rate' : float})
countries_data['Total tax rate'] = countries_data['Total tax rate']/100

countries_data['Unemployment rate'] = countries_data['Unemployment rate'].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'Unemployment rate' : float})
countries_data['Unemployment rate'] = countries_data['Unemployment rate']/100

countries_data['Gasoline Price'] = countries_data['Gasoline Price'].str[1:]
countries_data = countries_data.astype({'Gasoline Price' : float})

countries_data["GDP"] = countries_data['GDP'].str[1:]
countries_data = countries_data.astype({'GDP' : float})

countries_data['Minimum wage'] = countries_data['Minimum wage'].str[1:]
countries_data = countries_data.astype({'Minimum wage' : float})

countries_data['Out of pocket health expenditure'] = countries_data['Out of pocket health expenditure'].str.replace(r"%", "", regex=True)
countries_data = countries_data.astype({'Out of pocket health expenditure' : float})
countries_data['Out of pocket health expenditure'] = countries_data['Out of pocket health expenditure']/100
```

```python
countries_data = countries_data.astype({'Land Area(Km2)' : float})
countries_data = countries_data.astype({'Co2-Emissions' : float})
countries_data = countries_data.astype({'CPI' : float})
countries_data = countries_data.astype({'Population' : float})
countries_data = countries_data.astype({'Urban_population' : float})
```
