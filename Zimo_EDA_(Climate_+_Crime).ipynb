```python
# Keep relevant columns
cleaned_global_terrorism = global_terrorism[['eventid', 'iyear', 'imonth', 'country_txt', 'city', 'latitude', 'longitude', 'gname', 'nkill', 'nwound']]

# Replace Unknown with NaN
cleaned_global_terrorism.replace("Unknown", np.nan, inplace=True)

# Change country_txt to country
cleaned_global_terrorism.rename(columns={'country_txt': 'country', 'iyear' : 'year', 'imonth' : 'month'}, inplace=True)
```

```python
# Concatenate crime_40_60, crime_60_100, crime_100_250, crime_250_plus into one table
us_crime = pd.concat([crime_40_60, crime_60_100, crime_100_250, crime_250_plus], ignore_index=True)
us_crime
```

```python
null_counts = us_crime.isnull().sum()
null_counts
```

```python
# Drop 'tot_violent_crime' and 'tot_prop_crim' columns
cleaned_us_crime = us_crime.drop(['tot_violent_crime', 'tot_prop_crim'], axis=1)

# Calculate the total crime
crime_columns = ['violent_crime', 'murder', 'rape', 'robbery', 'agrv_assault', 'prop_crime', 'burglary', 'larceny', 'vehicle_theft', 'arson']
for col in crime_columns:
    cleaned_us_crime[col] = pd.to_numeric(cleaned_us_crime[col], errors='coerce')

cleaned_us_crime['total_crime'] = cleaned_us_crime[crime_columns].sum(axis=1, skipna=True)

# Rename columns
cleaned_us_crime.rename(columns={'cities': 'city', 'states': 'state'}, inplace=True)
cleaned_us_crime = cleaned_us_crime .drop(['state'], axis=1)
```

```python
# Convert to int type
cleaned_us_crime['population'] = cleaned_us_crime['population'].replace(',', '', regex=True).astype(int)
```

```python
# Split 'City' into 'City' and 'Country'
cleaned_city_crime_index = city_crime_index.copy()
city_country_split = cleaned_city_crime_index['City'].str.split(',', n=1, expand=True)
cleaned_city_crime_index['City'] = city_country_split[0].str.strip()
cleaned_city_crime_index['Country'] = city_country_split[1].str.strip().fillna('null')

# Further split the 'Country' column into 'State' and 'Country'
def split_state_country(value):
    if pd.isnull(value):
        return pd.Series([None, value])
    parts = value.split(',')
    if len(parts) > 1 and len(parts[0]) == 2 and parts[0].isupper():
        return pd.Series([parts[0], ' '.join(parts[1:])])
    else:
        return pd.Series([None, value])

cleaned_city_crime_index[['State', 'Country']] = cleaned_city_crime_index['Country'].apply(split_state_country)
```

```python
cleaned_city_crime_index['Country'] = cleaned_city_crime_index['Country'].str.lstrip()
```
